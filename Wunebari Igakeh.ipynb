#### Merging the dataset

Joined all four tables into one single dataframe

```python
df = order_details.merge(orders, on = "order_id", how = "inner").merge(pizza, on = "pizza_id", how = "inner").merge(pizza_types, on = "pizza_type_id", how = "inner")

df
```

#### Data cleaning

Convert Date and time from object to datetime

```python
#change the Dtype for date and time from object to Datatime

df["date"] = pd.to_datetime(df["date"])
df["time"] = pd.to_datetime(df["time"])
```

### Analysis Section

**1. What is the total revenue/sales ?**

Calculated the total revenue by multiplying quantity by price amd summing across all orders

```python
revenue = df["revenue"] = df["price"] * df["quantity"]  #creating a column for revenue
revenue
```

```python
df.head()
```

```python
total_revenue = df["revenue"].sum()
total_revenue
```

The total revenue/sales generated is 817860.05 

**2. Find the total quantity sold?**

Calculated the total number of pizza sold by summing the quantity column across all tranactions

```python
total_quantity_sold = df["quantity"].sum()
total_quantity_sold
```

**3. Find the total orders.**

Calculated the total number of orders by counting the unique order_id values

**4. How many pizza types do they sell?**

**5. Find the average price of the pizzas**

Calculated the average price by taking the mean of the price column across all records


```python
average_price = df["price"].mean()
average_price
```

The average price of pizza is 16.49

**6. What are the peak hours of sales?**

Analyse the peak hour of sales by extracting the hour from the time column and counting the number of unique orders per hour

```python
df["hour"] = df["time"].dt.hour
df["hour"]
```

```python
orders_per_hour = df.groupby("hour")["order_id"].nunique()
print(orders_per_hour)
```

```python
peak_hour = orders_per_hour.idxmax()
print("Peak Hour:", peak_hour)
```

**7. Find the total sales made on each day of the week. Which day of the week is when sales are made the most?**

We extract the day of the week, group the data by day and sum revenue to get sales per day then, find the day with the highest sales
 

```python
df["day_of_week"] = df["date"].dt.day_name()
```

```python
sales_per_day = df.groupby("day_of_week")["revenue"].sum()
sales_per_day
```

```python
best_day = sales_per_day.idxmax()
print("Highest Sales Day:", best_day)
```

**8. Find the top 5 bestselling pizzas.**

It was determined by grouping the data by pizza name, summing the total quantity sold for each, sorting in descending order and selecting the top five



```python
top_pizzas = df.groupby("name")["quantity"].sum().sort_values(ascending=False)
```

```python
top_5_pizzas = top_pizzas.head(5)
top_5_pizzas
```

**The top 5 bestselling pizzas are:**

The Classic Deluxe Pizza      
The Barbecue Chicken Pizza    
The Hawaiian Pizza            
The Pepperoni Pizza           
The Thai Chicken Pizza        

**9. Find the sales made in each month. Any trend noticeable?**


```python
df["month"] = df["date"].dt.month_name()
```

```python
monthly_sales = df.groupby("month")["revenue"].sum()
monthly_sales
```

```python
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_sales = monthly_sales.reindex(month_order)
print(monthly_sales)
```

```python
monthly_sales.plot(kind="bar", figsize=(10,5), color="blue", title="Monthly Pizza Sales")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.show()
```

From the bar chart plotted above, there is no noticeable trend.
